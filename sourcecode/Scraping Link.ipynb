```
finalLink = []
judulTresh = ["jokowi", "prabowo", "ma'ruf", "maruf", "amin", "sandiaga", "uno", "bpn", "tkn", "quick", "count", "kampanye", "gerindra", "pdip", "hitung", "cepat", "suara", "kalah", "menang", "coblos", "pemilu", "voting"]
judulPrevent = ["quick", "count", "hitung", "cepat"]
```

#DETIK.COM

##Mengambil Link Page Berita

```
linksDetik = []
searchQueries = ['Jokowi', 'Prabowo']

for keyword in searchQueries:
  links = []
  indexPage = 1
  while True:
    browser.get("https://www.detik.com/search/searchall?query=" + str(keyword) + "&siteid=2&sortby=time&fromdatex=17/11/2018&todatex=10/04/2019&page=" + str(indexPage))
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    #divs = soup.find(class_="list media_rows list-berita")
    
    for divs in soup.find_all("article", class_=False):
      for link in divs.find_all("a", href=True):
        get = link['href'].split('/')
        #print(get)
        for judul in judulTresh:
          if judul in get[5] and "berita" == get[3]:
            links.append(link['href'])
            break

    indexPage += 1
    if len(links) >= 20:
      linksDetik.append(links)
      break

#print(len(linksDetik[0]))
#print(len(linksDetik[1]))
#linksDetik
```

##Cleaning Hasil Link

```
flatLinks = [j for sub in linksDetik for j in sub]
#print(len(flatLinks))

uniqueLinks = [] 
for i in flatLinks: 
    if i not in uniqueLinks: 
        uniqueLinks.append(i) 

#print(len(uniqueLinks))
#uniqueLinks
```

```
if len(uniqueLinks) > 25:
  linkDetik = rd.sample(uniqueLinks, (rd.randint(20,25)))

finalLink.append(linkDetik)
print(len(linkDetik))
linkDetik
```

#LIPUTAN6.COM

##Mengambil Link Page Berita

```
linksLiputan6 = []
searchQueries = ['Jokowi', 'Prabowo']
#searchQueries = ['Jokowi']

for keyword in searchQueries:
  links = []
  indexPage = 1
  browser.get("https://www.liputan6.com/search?order=latest&channel_id=&from_date=17%2F11%2F2018&to_date=11%2F04%2F2019&type=all&q=" + str(keyword))
  html = browser.page_source
  soup = BeautifulSoup(html, "html.parser")

  for divs in soup.find_all("div", class_="articles--iridescent-list"):
    for articles in divs.find_all("article", class_="articles--iridescent-list--item articles--iridescent-list--text-item"):
      headerTitle = articles.find(class_="articles--iridescent-list--text-item__title")
      for link in headerTitle.find_all("a", href=True):
        get = link['href'].split('/')
        inputLink = True
        for judul in judulTresh:
          if judul in get[len(get)-1] and "pilpres" == get[3] and not "hot.liputan6.com" == get[2]:
            for prevent in judulPrevent:
              if prevent in get[len(get)-1]:
                inputLink = False
                break
            if inputLink:
              links.append(link['href'])
              break

  linksLiputan6.append(links)

#print(len(linksLiputan6[0]))
#print(len(linksLiputan6[1]))
#linksLiputan6
```

#INIPASTI.COM

##Mengambil Link Page Berita

```
linksIniPasti = []
searchQueries = ['Jokowi', 'Prabowo']

for keyword in searchQueries:
  links = []
  indexPage = 1
  
  while True:
    browser.get("https://inipasti.com/page/" + str(indexPage) + "/?s=" + str(keyword))
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    page = soup.find(class_="td-ss-main-content")
    date = page.find(class_="entry-date updated td-module-date")
    
    treshold = str(date.text)
    treshold = treshold.replace(',', '')
    split = treshold.split()
    split = split[:3]

    if (split[0] == 'April' and int(split[1]) <= 17 and split[2] == '2019'):
      break
    elif (split[0] == 'April' and split[2] == '2019'):
      indexPage += 1
    else:
      indexPage += 3

  while True:
    browser.get("https://inipasti.com/page/" + str(indexPage) + "/?s=" + str(keyword))
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    for divs in soup.find_all("div", class_="td_module_16 td_module_wrap td-animation-stack"):
      for headers in divs.find_all("h3", class_="entry-title td-module-title"):
        for link in headers.find_all("a", href=True):
          get = link['href'].split('/')
          for judul in judulTresh:
            if judul in get[3]:
              links.append(link['href'])
    
    indexPage += 1
    if len(links) >= 20:
      linksIniPasti.append(links)
      break

#print(len(linksIniPasti[0]))
#print(len(linksIniPasti[1]))
#linksIniPasti
```

#CHECK DUPLIKASI

```
def checkIfDuplicates_2(listOfElems):
  #''' Check if given list contains any duplicates '''    
  setOfElems = set()
  for elem in listOfElems:
    if elem in setOfElems:
      return True
    else:
      setOfElems.add(elem)         
      return False
```
